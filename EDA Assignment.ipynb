```python
#categorical variable analysis

def univariate(data,column,target):
    
    plt.figure(figsize=[10,8])
    plt.subplot(1, 2, 1)
    sns.countplot(data=data, x=column, hue='TARGET')
    plt.title(f'Count Plot of {column} with respect to TARGET')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=90)

    defaulters = data.groupby(column)['TARGET'].mean() * 100

    
    plt.subplot(1, 2, 2)
    defaulters.plot(kind='bar', color='c')
    plt.title(f' Defaulters within {column}')
    plt.xlabel(column)
    plt.ylabel('Percentage of Defaulters')
    plt.xticks(rotation=90)

    plt.tight_layout()
    plt.show()
```

```python
#categorical bivariate analysis
def bivariate(x,y,data,hue,labels):
    
    sns.barplot(x=x, y=y, data=data, hue=hue)
    plt.legend(labels=labels)
    plt.xticks(rotation=90)
    plt.show()
```

```python
#calling function
bivariate("NAME_INCOME_TYPE","AMT_INCOME_TOTAL",df,"TARGET",['Repayer','Defaulter'])
```

```python
#numerical univariate analysis
amount = df[['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE']]


fig = plt.figure(figsize=(10, 8))

for i, col in enumerate(amount):
    plt.subplot(2,2,i+1)
    sns.kdeplot(df.loc[df['TARGET'] == 1, col], label="Defaulter")
    sns.kdeplot(df.loc[df['TARGET'] == 0, col], label="Repayer")

    plt.legend()

plt.show()
```

```python
#numerical Byvariate
def bivariate(x,y,df,hue,kind,labels):
    plt.figure(figsize=[15,15])
    sns.relplot(x=x, y=y, data=df, hue=hue,kind=kind, legend = False)
    plt.legend(labels=labels)
    
    plt.show()
```

```python
bivariate('AMT_GOODS_PRICE','AMT_CREDIT',df,"TARGET" ,'line',['Repayer','Defaulter'])
```
